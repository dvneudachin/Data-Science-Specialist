## Предобработка данных

- Необходимо понять происхождение пропусков и при необходимости их заполнить;
- Сделаем проверку данных на наличие пропусков методом .isnull();
- Подсчитаем количество для каждого столбца, применив метод .sum().


```python
flats_data.isnull().sum()
```

С помощью цикла переберем значения из списка missing_values и сделаем проверку на уникальность значения.

```python
missing_values = ['balcony','is_apartment','parks_around3000','ponds_around3000']
for item in missing_values:
    print(flats_data[item].value_counts(), '\n')
```

Предположим, что в is_apartment пропуски означают, что квартиры не являются аппартаментами,  поэтому заполним на False.

Вероятно, пропуски в столбцах balcony, parks_around3000, ponds_around3000 указывают на отсутствие, значит укажем 0.






```python
flats_data['balcony'] = flats_data['balcony'].fillna(0)
flats_data['is_apartment'] = flats_data['is_apartment'].fillna(False)
flats_data['parks_around3000'] = flats_data['parks_around3000'].fillna(0)
flats_data['ponds_around3000'] = flats_data['ponds_around3000'].fillna(0)
```

Методом .unique() проверим столбец rooms на уникальность:

```python
number_of_rooms = flats_data['rooms'].unique()
number_of_rooms
```

Видим наличие квартир с числом комнат 0. Предположим, что это студии, т.е. однокомнатные квартиры маленькой площади.
Поэтому приравняем их к однокомнатным.

```python
flats_data['rooms'] = flats_data['rooms'].replace(0, 1)
```

С помощью метода .describe() проанализируем столбцы floors_total и ceiling_height:

```python
flats_data['ceiling_height'].describe()
```

Видим присутствие аномальных значений: минимальная высота потолка составляет 1 метр, а максимальная - 100 метров.
Заполним их медианным значением. 
Вероятно, природа пропусков обусловлена незнанием продавца о высоте потолка в квартире.

Найдем значение моды для значений высоты потолка.

```python
mode_ceiling = flats_data['ceiling_height'].mode()
mode_ceiling
```

Произведем замену пропусков на значение моды для значений высоты потолка менее 2.5 метра и более 5.

```python
flats_data.loc[flats_data['ceiling_height'] < 2.50, 'ceiling_height'] = mode_ceiling
flats_data.loc[flats_data['ceiling_height'] > 5, 'ceiling_height'] = mode_ceiling
```

```python
flats_data['ceiling_height'] = flats_data['ceiling_height'].fillna(2.5)
```

Посмотрим, сколько квартир имеет то или иное количество этажей:

```python
flats_data['floors_total'].value_counts()
```

Самое высокое жилое здание Санкт-Петербурга состоит из 37 этажей. 
Видим, что доля с количеством этажей более 37 составляет мизерную часть.

Проверим на наличие пропущенных значений стобцец locality_name

```python
flats_data['locality_name'].isnull().sum()
```

Пропущенные значения составляют незначительную часть. Тем более, не получится определить неизвестный населенный пункт.
Поэтому заменим пропуски словом unknown и проверим результат:

```python
flats_data['locality_name'] = flats_data['locality_name'].fillna('unknown')
flats_data['locality_name'].isna().sum()
```

Как упоминалось выше, названия населенных пунктов дублируются из-за лишних слов.
Уберем слова, которые затрудняют анализ ("деревня", "посёлок"):


```python
for i in range(0, 4):
    flats_data.loc[:, 'locality_name'] = flats_data['locality_name'].str.replace('^[а-я][а-яё]+\s', '')
```

Для населенных пунктов создадим новую переменную:

```python
locality_name_changed = flats_data['locality_name'].unique()
locality_name_changed
```

Теперь названия населенных пунктов более удобны для дальнейшего анализа.

Проверим на наличие пропущенных значений столбцы living_area и 'kitchen_area':


```python
flats_data['living_area'].isnull().sum()
```

```python
flats_data['kitchen_area'].isnull().sum()
```

Категориально по количеству комнат осуществим замены пропусков в столбцах living_area и kitchen_area. 
Для этого напишем цикл. Произведя замену на медианное значение, может выйти, что жилая площадь и площадь кухни больше, чем 
общая площадь квартиры. Необходимо сделать проверку с помощью условия living_area > total_area и kitchen_area > total_area,
а затем вывести количество строк после фильтрации методом .shape.

Вероятно, природа этих пропусков объясняется тем, что продавец не уделил должного внимания жилой площади и площади кухни,
т.к. в большинстве случаев всех интересует только общая площадь.

```python
for item in flats_data['rooms'].unique():
    filter_flags = flats_data['living_area'].isna() & (flats_data['rooms'] == item)
    flats_data.loc[filter_flags, 'living_area'] = flats_data.loc[flats_data['rooms'] == item, 'living_area'].median()
```

```python
for item in flats_data['rooms'].unique():
    filter_flags = flats_data['kitchen_area'].isna() & (flats_data['rooms'] == item)
    flats_data.loc[filter_flags, 'kitchen_area'] = flats_data.loc[flats_data['rooms'] == item, 'kitchen_area'].median()
```

```python
flats_data.query('living_area > total_area').shape
```

```python
flats_data.query('kitchen_area > total_area').shape
```

```python
flats_data.query('living_area + kitchen_area > total_area').shape
```

Видим, что 29 квартирах living_area > total_area. Заменим на разницу total_area и kitchen_area в этих квартирах. На результаты исследования это не повлияет.


```python
flats_data.loc[((flats_data['living_area'] + flats_data['kitchen_area']) > flats_data['total_area']), 'living_area'] = flats_data['total_area'] - flats_data['kitchen_area'] -0.0001
```

```python
flats_data.query('living_area + kitchen_area > total_area')
```

```python
flats_data['first_day_exposition'] = pd.to_datetime(flats_data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
flats_data['weekday'] = flats_data['first_day_exposition'].dt.weekday
flats_data['month'] = pd.DatetimeIndex(flats_data['first_day_exposition']).month
flats_data['year'] = flats_data['first_day_exposition'].dt.year
```

С помощью функции floor_type определим этаж квартиры. Категоризируем на варианты — first (первый), last (последний) и other (другой).

```python
def floor_type(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'first'
    elif floor ==  floors_total:
        return 'last'
    else:
        return 'other'
flats_data['floor_type'] = flats_data.apply(floor_type, axis= 1)
```



Необходимы значения диаграммы размаха для требуемого столбца. Для этого создадим функцию boxplot_values.



```python
def boxplot_values(column):
    upper_quartile = flats_data[column].quantile(.75)
    lower_quartile = flats_data[column].quantile(.25)
    iqr = upper_quartile - lower_quartile
    upper_whisker = flats_data[column][flats_data[column] <= upper_quartile + 1.5 * iqr].max()
    lower_whisker = flats_data[column][flats_data[column] >= lower_quartile - 1.5 * iqr].min()
    print('Нижний квартиль:', lower_quartile)
    print('Верхний квартиль:', upper_quartile)
    print('Нижний "ус":', lower_whisker)
    print('Верхний "ус":', upper_whisker)
```

Построим гистограмму и диаграмму размаха, которые покажут значения диаграммы размаха для требуемого столбца. 
Для этого напишем функцию hist_boxplot.




```python
def hist_boxplot(column):
    flats_data[column].hist(bins=200, figsize=(15,5)).set(xlabel=column, ylabel='Частота');
    plt.show()
    flats_data.boxplot(column=column, vert=False, figsize=(15,3)).set(xlabel='Частота');
    boxplot_values(column)
```
